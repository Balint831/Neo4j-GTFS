```python
import pandas as pd
pd.set_option('display.max_rows', 500)

pathways = pd.read_csv("data/pathways.txt")
stops = pd.read_csv("data/stops.txt")

stops_subdf = stops[["stop_id", "stop_name"]]
```

```python
path_stop_names = pathways.merge(stops_subdf, how = "inner", right_on = "stop_id", left_on = "from_stop_id")
path_stop_names.merge(stops_subdf, how = "inner", right_on = "stop_id", left_on = "to_stop_id" )
```

```python
path_stop_names.pathway_mode.value_counts()
```

If we look at the stops that these pathways link together, we can see, that these pathways connect underground stops with ones on the surface.
